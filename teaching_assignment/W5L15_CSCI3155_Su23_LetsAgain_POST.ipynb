# Ident and Let (`id` and `let id = e1 in e2`)

$$\begin{array}{c}
id \notin env
\\ \hline
eval(\; Ident(\; id\; ),\; env\; ) = ERROR
\end{array}\texttt{(ident-error)}$$
<br /><br /><br />

$$\begin{array}{c}
id \in env
,\;\;\;
env(id) = v
\\ \hline
eval(\; Ident(\; id\; ),\; env\; ) = v
\end{array}\texttt{(ident-ok)}$$
<br /><br /><br />

$$\begin{array}{c}
eval(e_1,\; env) = ERROR
\\ \hline
eval(\; Let(\; id,\; e_1,\; e_2\; ),\; env\; ) = ERROR
\end{array}\texttt{(let-error)}$$
<br /><br /><br />

$$\begin{array}{c}
eval(e_1,\; env) = v_1
,\;\;\;
v_1 \neq ERROR
,\;\;\;
\color{blue}{env \circ \{ id \mapsto v1 \} = env_{new}}
,\;\;\;
eval(e2,\; \color{red}{env_{new}}) = v_2
\\ \hline
eval(\; Let(\; id,\; e_1,\; e_2\; ),\; env\; ) = v_2
\end{array}\texttt{(let-ok)}$$

```scala
// LIVE extention! TDD~ish

type Environment = Map[String, Value]

sealed trait Expr {
    def eval(env: Environment): Value = this match {
        case Const(n) => NumValue(n)
        case Plus(e1, e2) => e1.eval(env) match {
            case NumValue(n1) => e2.eval(env) match {
                case NumValue(n2) => NumValue(n1 + n2)
                case _ => ErrorValue
            }
            case _ => ErrorValue
        }
        case Gt(e1, e2) => e1 eval env match {
            case NumValue(n1) => e2 eval env match {
                case NumValue(n2) => BoolValue(n1 > n2)
                case _ => ErrorValue
            }
            case _ => ErrorValue 
        }
        // HERE
        case Ident(id) => {
            if ( env contains id ) {
                env(id)
            } else {
                ErrorValue
            }
        }
        case Let(id, e1, e2) => e1 eval env match {
            case ErrorValue => ErrorValue
            case v1 => {
                val envNew = env + ( id -> v1 )
                val v2 = e2 eval envNew
                v2
            }
        }
        case _ => ???
    }
}
case class Const(f: Double) extends Expr
case class Plus(e1: Expr, e2: Expr) extends Expr
case class Gt(e1: Expr, e2: Expr) extends Expr
// HERE
case class Ident(id: String) extends Expr
// let id = e1 in e2
case class Let(id: String, e1: Expr, e2: Expr) extends Expr

sealed trait Value
case class NumValue(n: Double) extends Value
case class BoolValue(b: Boolean) extends Value
case object ErrorValue extends Value

val debug = true
def testEval(desc: String, e: Expr, vExpected: Value, env: Environment = Map()): Unit = {
    try {
        val vGot = e eval env
        if (vGot == vExpected) {
            println("$$ SUCCESS: " + desc)
        } else {
            println(s"XXXXXX FAILURE: $desc\n\tEXPR: $e\n\tEXPECTED: $vExpected\n\tFOUND: $vGot")
        }
        if (!debug) assert(vGot == vExpected)
    } catch {
        case err: Throwable => {
            println(s"XXXXXX FAILURE: $desc\n\tEXPR: $e\n\tEXPECTED: $vExpected\n\tFOUND: Something went horrably wrong $err")
            if (!debug) assert(false)
        }
    }
}

testEval("1", Const(1), NumValue(1))
testEval("2", Const(2), NumValue(2))
testEval("1 + 2", Plus(Const(1), Const(2)), NumValue(3))
testEval("3 + 4", Plus(Const(3), Const(4)), NumValue(7))
testEval("(1 + 2) + (3 + 4)", Plus(Plus(Const(1), Const(2)), Plus(Const(3), Const(4))), NumValue(10))
testEval("1 > 2", Gt(Const(1), Const(2)), BoolValue(false))
testEval("4 > 3", Gt(Const(4), Const(1)), BoolValue(true))
testEval("(1 + 2) > (3 + 4)", Gt(Plus(Const(1), Const(2)), Plus(Const(3), Const(4))), BoolValue(false))
testEval("100 > (3 + 4)", Gt(Const(100), Plus(Const(3), Const(4))), BoolValue(true))
testEval("1 + (3 > 2)", Plus(Const(1), Gt(Const(3), Const(2))), ErrorValue)
testEval("(2 > 3) + 4", Plus(Gt(Const(2), Const(3)), Const(4)), ErrorValue)

def testEval(desc: String, e: Expr, env: Environment, vExpected: Value): Unit = {
    testEval(desc, e, vExpected, env)
}
// HERE
testEval("x", Ident("x"), ErrorValue)
// env(id)
    // env: Map(("x" -> NumValue(2)))
    // id: "x"
    // result: NumValue(2)
testEval("x UNDER { x -> 2 }", Ident("x"), Map(("x" -> NumValue(2))), NumValue(2))
testEval("y UNDER { y -> 5 }", Ident("y"), Map(("y" -> NumValue(5))), NumValue(5))
testEval("z UNDER { z -> true }", Ident("z"), Map(("z" -> BoolValue(true))), BoolValue(true))
testEval("x UNDER { y -> 2 }", Ident("x"), Map(("y" -> NumValue(2))), ErrorValue)
testEval("let x = 1 in x", Let("x", Const(1), Ident("x")), Map(): Environment, NumValue(1))
testEval("let x = 2 in x", Let("x", Const(2), Ident("x")), NumValue(2))
testEval("let x = 2 in let y = 3 in x + y", Let("x", 
                                                Const(2), 
                                                Let("y", 
                                                    Const(3), 
                                                    Plus(Ident("x"), Ident("y")))), NumValue(5))
```

# IfThenElse `if (e1) e2 else e3`
* If-else `statement` vs `expression`
    * often if-else is a `statement` in that is has no value
    * But there is also the ternary expression that has a value
    * JAVA: `(e1) ? e2 : e3`
    * Python: `e2 if e1 else e3` (see in PS5)

# And `e1 && e2`
* a topic: **`short circuiting`**
* recall truth table
    * true && true = true
    * true && false = false
    * false && true = false
    * false && false = false
* generically
    * false && x = false
    * true && x = x
* so let us not look at `x` if we don't actually have to
    * has unintended consequences sometimes
    * Seen in C code to check that a pointer is non-null AND THEN increment the pointer
* **For LETTUCE**
    * only allow BoolValue on `&&`/`And(e1, e2)`...

$$\begin{array}{c}
eval(e_{1},\; env) = v_1
,\;\;\;
v_1 \notin \{ true,\; false \}
\\ \hline
eval(\; And(\; e_{1},\; e_{2}\; ),\; env\; ) = ERROR
